```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Aditi/My Work/Udemy/Finance Python/Housing.xlsx")
df
```

Independent variables : "House Size (sq.ft.)", "Number of Rooms", "Year of Construction"

```python
X = df.loc[ : , ["House Size (sq.ft.)", "Number of Rooms", "Year of Construction"]]
Y = df.loc[ :, ["House Price"]]
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y,X1).fit()

reg.summary()
```
